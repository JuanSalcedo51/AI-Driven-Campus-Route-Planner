```python
#@title Make a cell to build the basic structure from JSON

import json
import math
import heapq
import networkx as nx
import osmnx as ox

settings = campus_data["settings"]
belief_model = campus_data["belief_model"]

entrances = campus_data["entrances"]
parking_lots = campus_data["parking_lots"]
destinations = campus_data["destinations"]

all_special_nodes = {}
all_special_nodes.update(entrances)
all_special_nodes.update(parking_lots)
all_special_nodes.update(destinations)

node_type = {
    name: data["type"]
    for name, data in all_special_nodes.items()
}

base_full_prob = {
    lot_name: lot_data["Base_full_probability"]
    for lot_name, lot_data in parking_lots.items()
}

label_to_key = {}
for key, data in all_special_nodes.items():
    label_to_key[key.lower()] = key
    label_to_key[data["label"].lower()] = key

def display_name(node):
    if node in all_special_nodes:
        return all_special_nodes[node]["label"]
    return f"Road node {node}"

def resolve_user_choice(user_text):
    key = label_to_key.get(user_text.strip().lower())
    if key is None:
        raise ValueError("Invalid name. Use one of the listed labels.")
    return key

# Make the graph using the the OSMnx library.
G = ox.graph_from_place(
    "Florida Atlantic University, Boca Raton, FL",
    network_type="drive"
)

# Connect the special nodes from the json file with the road nodes.
nearest_osm_nodes = {}
for node_name, data in all_special_nodes.items():
    nearest_osm_nodes[node_name] = ox.distance.nearest_nodes( # sanp the coordinates to closest valid road intersection in the OSMnx graph.
        G,
        X=data["lon"],
        Y=data["lat"]
    )

for node_name, data in all_special_nodes.items():
    nearest_node = nearest_osm_nodes[node_name]

    G.add_node(
        node_name,
        y=data["lat"],
        x=data["lon"],
        node_type=data["type"],
        label=data["label"]
    )

    # Make bidirectionals routes between the road nodes connecting with the lots nodes.
    G.add_edge(nearest_node, node_name, length=10, mode="access", weight=0) #weight is 0 because is just an access edge.
    G.add_edge(node_name, nearest_node, length=10, mode="access", weight=0)
# Real costs will be calculated later.
print(f"Graph ready: {len(G.nodes)} nodes, {len(G.edges)} edges")
```

```python
#@title Translate from user-input time to decimal number, for implement in the belief.

def parse_time_to_decimal(time_str):
    parts = time_str.strip().split(":")

    if len(parts) == 1:
        hour = int(parts[0])
        minutes = 0
    elif len(parts) == 2:
        hour = int(parts[0])
        minutes = int(parts[1])
    else:
        raise ValueError("Invalid time format. Use HH or HH:MM.")

    if not (0 <= hour <= 23 and 0 <= minutes <= 59):
        raise ValueError("Hour must be 0-23 and minutes 0-59.")

    return hour + minutes / 60.0


def get_parking_probability(hour, lot_name): # This functions works as a basic belief updating model.
    base_p = base_full_prob[lot_name] # take the base probability about full parking space and updating it based on hour.
    popularity = parking_lots[lot_name]["popularity"]

    peak_start = belief_model["peak_hours"]["start"]
    peak_end = belief_model["peak_hours"]["end"]
    post_peak_end = belief_model["post_peak_end"]

    peak_increase = belief_model["peak_increase"][popularity]
    midday_increase = belief_model["midday_increase"][popularity]
    afternoon_decrease = belief_model["afternoon_decrease"][popularity]

    p = base_p

# Update the beliefs based on the time of the day.
    if peak_start <= hour <= peak_end:
        p += peak_increase
    elif peak_end < hour < post_peak_end:
        p += midday_increase
    elif hour >= post_peak_end:
        p -= afternoon_decrease

    return max(0.0, min(1.0, p))


def availability_penalty(prob_full): # Translate the probabilistic beliefs into a concrete cost for A*
    safe_limit = belief_model["penalty_thresholds"]["safe_limit"]
    risk_limit = belief_model["penalty_thresholds"]["risk_limit"]
    medium_penalty = belief_model["penalty_thresholds"]["medium_penalty"]
    high_penalty = belief_model["penalty_thresholds"]["high_penalty"] # IF the risk is high for example, a heavy time penalty will be added.
# The agent will based its decision if choose a parking lot or not.
    if prob_full < safe_limit:
        return 0.0
    elif prob_full < risk_limit:
        return medium_penalty
    else:
        return high_penalty


def print_lot_beliefs(hour):
    print(f"\nBeliefs for {hour:.2f}:")
    for lot_name in parking_lots:
        p = get_parking_probability(hour, lot_name)
        penalty = availability_penalty(p)
        print(f"{display_name(lot_name)}: P(full) = {p:.2f}, penalty = {penalty}")
```

```python
#@title Build a search graph from the previus combined graph (known as G)

'''
The main idea to build another graph is for avoid routes in which the vehicles would "enter" to a building from the road network,
instead, the building only will be reach from a parking lot across a walking transition.
'''
def meters_to_drive_minutes(meters, drive_speed_mph):
    miles = meters / 1609.34
    hours = miles / drive_speed_mph
    return hours * 60.0


def meters_to_walk_minutes(meters, walk_speed_mph):
    miles = meters / 1609.34
    hours = miles / walk_speed_mph
    return hours * 60.0

# The following function represent a Haversine formula to modelate straight distances based on coordinates and the curved representation of the earth
def haversine_meters(lat1, lon1, lat2, lon2):
    R = 6371000 # Earth Radius.
    phi1 = math.radians(lat1)
    phi2 = math.radians(lat2)
    dphi = math.radians(lat2 - lat1)
    dlambda = math.radians(lon2 - lon1)

#Haversine Formula, suggested by GenAI because coordinates were used for this project.
    a = math.sin(dphi / 2) ** 2 + math.cos(phi1) * math.cos(phi2) * math.sin(dlambda / 2) ** 2
    return 2 * R * math.atan2(math.sqrt(a), math.sqrt(1 - a))


def build_search_graph(G, hour):
    search_graph = nx.DiGraph()

    lot_names = set(parking_lots.keys())
    destination_names = set(destinations.keys())

    for n, data in G.nodes(data=True):
        search_graph.add_node(n, **data)

    for u, v, data in G.edges(data=True): # Prevent the agent for "driving" directly to the destination building
        if u in destination_names or v in destination_names:
            continue # The idea is stop in a parking lot first and later walk to the building.

        if u in lot_names:
            continue

        length_m = data.get("length", 1.0)
        drive_minutes = meters_to_drive_minutes(length_m, settings["drive_speed_mph"])

        if search_graph.has_edge(u, v):
            if drive_minutes < search_graph[u][v]["weight"]:
                search_graph[u][v]["weight"] = drive_minutes
                search_graph[u][v]["mode"] = "drive"
        else:
            search_graph.add_edge(u, v, weight=drive_minutes, mode="drive")

    for lot_name, lot_data in parking_lots.items(): # Get the belief and penalty for this specific hour.
        p_full = get_parking_probability(hour, lot_name)
        penalty = availability_penalty(p_full)

        for dest_name, dest_data in destinations.items(): # Calculate physical walking distance.
            walk_m = haversine_meters(
                lot_data["lat"], lot_data["lon"],
                dest_data["lat"], dest_data["lon"]
            )

            #Multiply the walk_wight to model human preferences (driving instead of walking)
            # Add the probabilistic penalty if the lot would be full.
            walk_minutes = meters_to_walk_minutes(walk_m, settings["walk_speed_mph"])
            total_walk_cost = walk_minutes * settings["walk_weight"] + penalty

            search_graph.add_edge(
                lot_name,
                dest_name,
                weight=total_walk_cost, # This abstract cost is what A* will try to minimize.
                mode="walk",
                prob_full=p_full,
                penalty=penalty,
                walk_minutes=walk_minutes # Kept strictily for user interface.
            )

    return search_graph
```

```python
#@title The A* Algorithm for the search

def get_node_coords(search_graph, node):
    data = search_graph.nodes[node]
    if "y" in data and "x" in data:
        return data["y"], data["x"]
    return None


def heuristic(search_graph, node, goal): # Represent h(n) for A*
# It calculates the straight line distance from current node to the goal.
# it will never overestimate because the straight line is the shortest path.
    if node == goal:
        return 0.0

    node_coords = get_node_coords(search_graph, node)
    goal_coords = get_node_coords(search_graph, goal)

    if node_coords is None or goal_coords is None:
        return 0.0

    straight_m = haversine_meters(
        node_coords[0], node_coords[1],
        goal_coords[0], goal_coords[1]
    )

# If the transition is walking after parking, apply the walk_wight to keep h(n) consistent with g(n).
    if node in parking_lots and goal in destinations:
        return meters_to_walk_minutes(
            straight_m,
            settings["walk_speed_mph"]
        ) * settings["walk_weight"]

    return meters_to_drive_minutes(
        straight_m,
        settings["drive_speed_mph"]
    )


def A_star_campus(search_graph, start, goal): # Standard A* implementation using priority queue. Minimizing f(n) = g(n) + h(n)
    frontier = [(heuristic(search_graph, start, goal), 0.0, start, [start])]
    best_cost = {start: 0.0}

    while frontier:
        f_cost, g_cost, current, path = heapq.heappop(frontier)

        if current == goal:
            return path, g_cost

        if g_cost > best_cost.get(current, float("inf")):
            continue

        for neighbor in search_graph.neighbors(current):
            edge_weight = search_graph[current][neighbor]["weight"]
            new_g = g_cost + edge_weight

            if new_g < best_cost.get(neighbor, float("inf")):
                best_cost[neighbor] = new_g
                new_f = new_g + heuristic(search_graph, neighbor, goal)
                heapq.heappush(frontier, (new_f, new_g, neighbor, path + [neighbor]))

    return None, float("inf")
```

```python
#@title Make a function to execute all together and show the selected parking lot.

def extract_selected_lot(path):
    lots_in_path = [node for node in path if node in parking_lots]
    return lots_in_path[-1] if lots_in_path else None

def get_cost_breakdown(path, search_graph):
    drive_cost = 0.0
    walk_time_physical = 0.0

    for u, v in zip(path, path[1:]):
        edge = search_graph[u][v]
        if edge.get("mode") == "drive":
            drive_cost += edge["weight"]
        elif edge.get("mode") == "walk":
            walk_time_physical += edge["walk_minutes"]

    return drive_cost, walk_time_physical

def run_route(start_input, goal_input, time_str):
    start = resolve_user_choice(start_input)
    goal = resolve_user_choice(goal_input)

    hour = parse_time_to_decimal(time_str)

    # Search graph and A* execution
    search_graph = build_search_graph(G, hour)
    path, cost = A_star_campus(search_graph, start, goal)
    selected_lot = extract_selected_lot(path) if path else None

    # User Interface.
    print("\n" + "="*50)
    print("      CAMPUS ERRAND ROUTE PLANNER      ")
    print("="*50)

    if path and selected_lot:
        drive_cost, walk_time_physical = get_cost_breakdown(path, search_graph)
        p = get_parking_probability(hour, selected_lot)
        penalty = availability_penalty(p)

        print(f"Origin:             {display_name(start)}")
        print(f"Destination:        {display_name(goal)}")
        print(f"Time of arrival:    {time_str}")
        print("-" * 50)
        print("AGENT DECISION:")
        print(f"Recommended Lot:    {display_name(selected_lot)}")
        print(f"Full Probability:   {p:.0%}")
        print(f"Risk Penalty:       {penalty:.2f} cost units")
        print("-" * 50)
        print("ROUTE METRICS:")
        print(f"Driving Time:       {drive_cost:.2f} minutes")
        print(f"Walking Time:       {walk_time_physical:.2f} minutes")
        print(f"Total Path Cost:    {cost:.2f} units (Abstract)")
    else:
        print("Error: No valid path could be found for the specified parameters.")

    print("="*50 + "\n")

    return path, cost, selected_lot

def diagnose_lot_costs(start_input, goal_input, time_str):
    start = resolve_user_choice(start_input)
    goal = resolve_user_choice(goal_input)

    hour = parse_time_to_decimal(time_str)
    search_graph = build_search_graph(G, hour)

    rows = []

    for lot_name in parking_lots:
        try:
            drive_cost = nx.shortest_path_length(search_graph, start, lot_name, weight="weight")
            walk_edge = search_graph[lot_name][goal]
            walk_cost = walk_edge["weight"]
            penalty = walk_edge["penalty"]
            p_full = walk_edge["prob_full"]
            total = drive_cost + walk_cost

            rows.append({
                "lot": display_name(lot_name),
                "drive_cost": round(drive_cost, 2),
                "walk_cost": round(walk_cost, 2),
                "penalty": round(penalty, 2),
                "p_full": round(p_full, 2),
                "total_cost": round(total, 2)
            })
        except Exception:
            pass

    rows = sorted(rows, key=lambda x: x["total_cost"])

    print(f"\nCost diagnosis from {display_name(start)} to {display_name(goal)} at {time_str}:")
    for row in rows:
        print(row)
```
